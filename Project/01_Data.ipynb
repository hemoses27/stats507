# Project Data

**Goal**: Create a training and testing dataset to be used for modeling  
**Inputs**: AutoInsurance.csv  
**Outputs**: test.csv and train.csv

```python
# load necessary packages
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# load dataset
car = pd.read_csv("AutoInsurance.csv")
```

```python
# split data into train and test set
train, test = train_test_split(car, test_size=0.2, random_state=1220)
```
