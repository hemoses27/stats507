# Project Linear Model

**Goal**: Train the best multiple linear regression model possible and evaluate its performance  
**Inputs**: train.csv and test.csv  
**Outputs**: fig1.png

```python
# load necessary packages
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load datasets
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

## Model Building

```python
# fit initial regression model
model = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Policy Type') + Q('Vehicle Size') + State + Q('Number of Policies') + Q('Income') + Q('Months Since Last Claim') + Q('Months Since Policy Inception') + Q('Number of Open Complaints') + Q('Total Claim Amount') + Gender + Q('Marital Status')", data=train).fit()

# view model results
print(model.summary())
```

```python
# fit regression model without insignificant predictors
model2 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Vehicle Size') + Q('Income') + Q('Months Since Policy Inception') + Q('Total Claim Amount') + Gender + Q('Marital Status')", data=train).fit()

# view model results
print(model2.summary())
```

```python
# add polinomial term for income
model4 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + Gender + Q('Marital Status')", data=train).fit()

# view model results
print(model4.summary())
```

```python
# add polinomial term for total claim amount
model5 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + I(Q('Total Claim Amount')**2) + Gender + Q('Marital Status')", data=train).fit()

# view model results
print(model5.summary())
```

```python
# add polinomial term for total claim amount
model6 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + I(Q('Total Claim Amount')**2) + I(Q('Total Claim Amount')**3) + Gender + Q('Marital Status')", data=train).fit()

#  view model results
print(model6.summary())
```

```python
# add polinomial term for total claim amount
model6 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + I(Q('Total Claim Amount')**2) + I(Q('Total Claim Amount')**3) + I(Q('Total Claim Amount')**4) + Gender + Q('Marital Status')", data=train).fit()

# View model results
print(model6.summary())
```

```python
# add polinomial term to cutomer lifetime value
model7 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + I(Q('Customer Lifetime Value')**2) + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + I(Q('Total Claim Amount')**2) + I(Q('Total Claim Amount')**3) + I(Q('Total Claim Amount')**4) + Gender + Q('Marital Status')", data=train).fit()

# view model results
print(model7.summary())
```

```python
# add polinomial term to customer lifetime value
model8 = smf.ols("Q('Monthly Premium Auto') ~ Q('Customer Lifetime Value') + I(Q('Customer Lifetime Value')**2) + I(Q('Customer Lifetime Value')**3) + Education + Q('Vehicle Size') + Income + I(Income**2) + Q('Months Since Policy Inception') + Q('Total Claim Amount') + I(Q('Total Claim Amount')**2) + I(Q('Total Claim Amount')**3) + I(Q('Total Claim Amount')**4) + Gender + Q('Marital Status')", data=train).fit()

# View model results
print(model8.summary())
```

## Performance Metrics

```python
# add model predictions to test dataset
test['Predicted Monthly Premium Auto'] = model8.predict(test)
```

```python
# calculate R^2 for test dataset
r2_test = r2_score(test["Monthly Premium Auto"], test["Predicted Monthly Premium Auto"])
r2_test
```

```python
# calculate training data rmse
rmse = np.sqrt(model.mse_resid)
print("Train RMSE:", rmse)
```

```python
# calculate testing data rmse
rmse_test = np.sqrt(np.mean((test["Monthly Premium Auto"] - test["Predicted Monthly Premium Auto"])**2))

print("Test RMSE:", rmse_test)
```

```python
# create scatter plot
sns.scatterplot(data=test, x="Monthly Premium Auto", y="Predicted Monthly Premium Auto")

# add y = x line through origin
lims = [
    0,  
    max(test["Predicted Monthly Premium Auto"].max(), test["Monthly Premium Auto"].max())
]
plt.plot(lims, lims, 'r-', label='y = x')
plt.xlim(0, lims[1])
plt.ylim(0, lims[1])
plt.title("LM Prediction Performance")
plt.legend()
plt.savefig("fig1.png", dpi=300, bbox_inches="tight")
plt.show()
```
